## **Constraint Satisfaction Problems**




Install library constraint (jika belum ada)

```python
!pip install python-constraint
from constraint import Problem
```

Membuat objek masalah CSP, aturan wilayah

```python
problem = Problem()
# Daftar wilayah yang akan diwarnai
wilayah = ["A", "B", "C", "D", "E", "F"]
# Tambahkan variabel dengan domain warna yang tersedia
warna = ["Merah", "Kuning", "Hijau", "Orange"]


for w in wilayah:
  problem.addVariable(w, warna)

# Tambahkan constraint (wilayah yang berdekatan tidak boleh memiliki warna yang sama)
tetangga = [("A", "B"), ("A", "C"), ("B", "C"), ("B", "D"), ("C", "D"),("D","E"),("E","F")]
for (wil1, wil2) in tetangga:
  problem.addConstraint(lambda w1, w2: w1 != w2, (wil1, wil2))
```

Melakukan pencarian solusi dan menampilkan hasil


```python
solusi = problem.getSolutions()

# Cetak hasil solusi
print(f"Ditemukan {len(solusi)} solusi yang mungkin:")
for sol in solusi:
  print(sol)
```

## **Rule-based Artificial Intelligence**


Definisikan kelas diagnose penyakit

```python
def diagnose_penyakit(gejala):
    rules = {
        "demam": ["flu", "malaria"],
        "batuk": ["flu", "tuberkulosis"],
        "sakit kepala": ["flu", "migren"],
        "mual": ["malaria", "keracunan makanan"],
        "ruam kulit": ["campak", "alergi"]
        }
    kemungkinan_penyakit = {}

    for gejala_input in gejala:
        for penyakit, gejala_penyakit in rules.items():
            if gejala_input in rules:
                for p in rules[gejala_input]:
                    kemungkinan_penyakit[p] = kemungkinan_penyakit.get(p, 0) + 1

    if not kemungkinan_penyakit:
        return "Tidak ada penyakit yang cocok dengan gejala yang diberikan."

    # Mencari penyakit dengan kemungkinan yang tinggi
    diagnosis = sorted(kemungkinan_penyakit.items(), key=lambda x: x[1], reverse=True)

    return f"Diagnose yang memungkinkan:{diagnosis[0][0]} (kemungkinan {diagnosis[0][1]} gejala cocok)"
```

Input dari user

```python
# Contoh input user
gejala_user = ["batuk","sakit kepala", "demam"]
hasil_diagnose = diagnose_penyakit(gejala_user)
print(hasil_diagnose)
```

## Case robot

Install Pyhop yaitu library Python (otomatisasi perencanaan tugas)

```python
# !wget https://raw.githubusercontent.com/ShopRunner/pyhop/master/pyhop.py
!wget https://raw.githubusercontent.com/marcoparadina/pyhop/master/pyhop.py
import pyhop
```

Definisikan aksi

```python
def move(state,robot,loc1,loc2):
  if state.posisi[robot] == loc1:
    state.posisi[robot] = loc2
    return state
  return False

def pickup(state,robot,paket):
  if state.posisi[robot] == state.posisi[paket] and not state.paket_dipegang[robot]:
    state.paket_dipegang[robot] = paket
    state.posisi[paket] = "dipegang"
    return state
  return False

def drop(state,robot,paket):
  if state.paket_dipegang[robot] == paket:
    state.paket_dipegang[robot] = None
    state.posisi[paket] = state.posisi[robot]
    return state
  return False


pyhop.declare_operators(move,pickup,drop)
```

Definisikan tugas (tasks)

```python
def deliver(state,robot,paket,tujuan):
  if state.posisi[paket] == tujuan:
    return []
  return [
      ("move",robot,state.posisi[robot],state.posisi[paket]),
      ("pickup",robot,paket),
      ("move",robot,state.posisi[paket],tujuan),
      ("drop",robot,paket)
  ]

pyhop.declare_methods("deliver",deliver)
```

Definisikan kondisi awal

```python
# Buat state awal
state1 = pyhop.State("state1")
state1.posisi = {"robot1": "A", "paket1": "B"}
state1.paket_dipegang = {"robot1": None}
print("Keadaan awal: ", vars(state1))
```

```python
# Rencanakan solusi
plan = pyhop.pyhop(state1, [("deliver", "robot1", "paket1", "C")], verbose=1)
print("Rencana Aksi yang ditemukan:", plan)
```

# Latihan

Buat state awal untuk 2 robot

```python
# Buat state awal
state2 = pyhop.State("state2")

# Posisi awal robot dan paket
state2.posisi = {
    "robot1": "A",
    "robot2": "B",
    "paket1": "A",
    "paket2": "C"
}

# Robot belum pegang paket apa pun
state2.paket_dipegang = {
    "robot1": None,
    "robot2": None
}

print("Keadaan awal:", vars(state2))
```

```python
# Task: robot2 harus mengantar paket2 ke lokasi D
plan2 = pyhop.pyhop(state2, [("deliver", "robot2", "paket2", "D")], verbose=1)
print("Rencana aksi yang ditemukan:", plan2)
```
