Create Dataset Mahasiswa AI (Sederhana)

Looping

```python
np.random.seed(42)

# Data 30 mahasiswa praktikum AI (lebih sederhana)
data_mahasiswa = {
'Nama': [f'Mahasiswa_{j+1:02d}' for j in range(1, 31)], #j=1, j<31
  'Quiz1': np.random.randint(60, 80, 30),
  'Quiz2': np.random.randint(65, 95, 30),
    'Praktikum': np.random.randint(70, 100, 30),
    'Kehadiran': np.random.randint(80, 100, 30)
}

df_mahasiswa = pd.DataFrame(data_mahasiswa)
# menampilkan 5 data teratas
print(df_mahasiswa)

```

```python
df_mahasiswa.head()
```

Hitung nilai akhir sederhana

Bobot nilai :
- Quiz : 25%
- Praktikum : 40%
- Kehadiran : 10%

```python
df_mahasiswa['Nilai_Akhir'] = (df_mahasiswa['Quiz1'] * 0.25 +
                     df_mahasiswa['Quiz2'] * 0.25 +
                     df_mahasiswa['Praktikum'] * 0.40 +
                     df_mahasiswa['Kehadiran'] * 0.10)


print(f"Total mahasiswa: {len(df_mahasiswa)}")
print("\nPreview data:")
# print(df_mahasiswa)
df_mahasiswa.head(10)
```

Hitung nilai teratas

```python
print(f"Rata-rata nilai akhir: {df_mahasiswa['Nilai_Akhir'].mean():.2f}")
print(f"Nilai tertinggi: {df_mahasiswa['Nilai_Akhir'].max():.1f}")
print(f"Nilai terendah: {df_mahasiswa['Nilai_Akhir'].min():.3f}")

print("\nTOP 5 MAHASISWA:")
top_5 = df_mahasiswa.nlargest(5, 'Nilai_Akhir')[['Nama', 'Nilai_Akhir']]
print(top_5)

# OPTIONAL
print("\nRATA-RATA PER KOMPONEN:")
for col in ['Quiz1', 'Quiz2', 'Praktikum', 'Kehadiran']:
    print(f"   {col}: {df_mahasiswa[col].mean():.2f}")
# print()
```

Simple Visualization (2 plots saja untuk waktu)

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
#plt.subplots(baris, kolom, figsize = (panjang, lebar))

# Plot 1: Histogram Nilai Akhir
axes[0].hist(df_mahasiswa['Nilai_Akhir'], bins=8, color='blue', alpha=0.7, edgecolor='navy')
axes[0].set_title('Distribusi Nilai Akhir')
axes[0].set_xlabel('Nilai Akhir')
axes[0].set_ylabel('Jumlah Mahasiswa')
axes[0].grid(True, alpha=0.3)

# # Plot 2: Bar Chart Rata-rata Komponen
komponen = ['Quiz1', 'Quiz2', 'Praktikum', 'Kehadiran']
rata2 = [df_mahasiswa[k].mean() for k in komponen]
axes[1].bar(komponen, rata2, color='green')
axes[1].set_title('Rata-rata per Komponen')
axes[1].set_ylabel('Nilai Rata-rata')
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(False)

plt.tight_layout()
plt.show()
```
